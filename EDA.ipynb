```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("books_data.csv")

df.head()
```

```python
df.info()
```

```python
df["Price"] = df["Price"].str.replace("Â£", "", regex=False)

df["Price"] = df["Price"].astype(float)

df.head()
```

```python
df.describe()
```

```python
top_books = df.sort_values(by="Price", ascending=False)

top_books.head(10)
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Price"], bins=10)

plt.title("Book Price Distribution")

plt.show()
```

```python
top10 = df.sort_values(by="Price", ascending=False).head(10)

plt.figure(figsize=(10,5))

sns.barplot(data=top10, x="Price", y="Title")

plt.title("Top 10 Most Expensive Books")

plt.show()
```

```python
avg_price = df["Price"].mean()

print("Average Book Price:", round(avg_price, 2))
```

```python
cheapest = df.loc[df["Price"].idxmin()]

print(cheapest)
```

```python
expensive = df.loc[df["Price"].idxmax()]

print(expensive)
```

# Business Insights

1. Book prices vary significantly across the catalog.
2. Most books are concentrated in the mid-price range.
3. A few premium-priced books increase the overall average price.
4. Different pricing strategies appear to be used across books.

# Recommendations

1. Categorize books into Low, Medium, and Premium segments.
2. Focus promotions on mid-priced books since they form the largest group.
3. Analyze premium-priced books separately to understand high-value offerings.
4. Monitor pricing trends to improve sales strategy.
